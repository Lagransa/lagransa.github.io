```python
init_state = 0
trans_matrix = np.array([[0.9, 0.95, 1], [0.84, 0.8, 1], [0.75, 1, 0.5]])
def next_step(matrix, state):
    try:
        new_state = 0
        rand = random.random()
        trans_probability = matrix[state]
        sorted_idx = np.argsort(trans_probability)
        for idx in sorted_idx:
            if rand < trans_probability[idx]:
                new_state = idx
                break
        return new_state
    except Exception as e:
        print(e)

state = init_state
state_records = []

for i in range(10000):
    state = next_step(trans_matrix, state)
    state_records.append(state)

print(np.unique(state_records, return_counts=True))
```
